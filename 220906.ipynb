```python
n=np.random.random(100)
# print(n)
print("기본 함수")
print(sum(n))
print(min(n))
print(max(n))
print("\n넘파이 함수")
print(np.sum(n))
print(np.min(n))
print(np.max(n))
```

```python
print(np.mean(n))
print(np.median(n))
print(np.std(n))
print(np.var(n))
print(np.argmin(n), np.min(n))
print(np.argmax(n), np.max(n))
print(np.any(n))
print(np.all(n))
```
