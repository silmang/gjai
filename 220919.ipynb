```python
from collections import Counter
import matplotlib
from matplotlib import font_manager, rc
font_path = 'C:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rc('font', family=font_name)
import matplotlib.pyplot as plt

import rhinoMorph
rn = rhinoMorph.startRhino()
# morph = rhinoMorph.onlyMorph_list(rn, word_list, pos=['NNG', 'NNP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi=True)
```

```python
import pandas as pd
import warnings
warnings.filterwarnings("error")
path = "C:\\pytest\\\\"
```

```python
data = pd.read_csv(path+"()_Validation.csv")
data = data[['','','_','1','2','3','4']]
data
```

```python
data.info()
```

```python
temp_df1 = data[['','','_','1']].rename(columns={'1':''})
temp_df2 = data[['','','_','2']].rename(columns={'2':''})
temp_df3 = data[['','','_','3']].dropna().rename(columns={'3':''})
temp_df4 = data[['','','_','4']].dropna().rename(columns={'4':''})
```

```python
refined_data = pd.concat([temp_df1, temp_df2, temp_df3, temp_df4], axis=0)
refined_data.reset_index(inplace=True)
refined_data.drop('index', axis=1, inplace=True)
refined_data
```

```python
refined_data[refined_data['_']!='']
```

```python
refined_data.nunique()
```

```python
for i in refined_data.columns:
    print(i, refined_data[i].unique())
```

```python
stopwords_ko = ["", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", "", ""]
most_words = 30
custom_stopwords_ko = ['','','','','','','']

stopwords_ko = stopwords_ko+custom_stopwords_ko
result = pd.DataFrame()
for i in refined_data[''].unique():
    for j in refined_data[''].unique():
        for k in refined_data['_'].unique():
            temp_df = refined_data.loc[(refined_data['']==i) & (refined_data['']==j) & (refined_data['_']==k)]
            temp_sentences_list = temp_df[''].to_list()
            sentences = ' '.join(temp_sentences_list)
            morph = rhinoMorph.onlyMorph_list(rn, sentences, pos=['NNG', 'NNP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi=True)

            counter_morph = Counter(morph)
            counter_morph = sorted(counter_morph.items(), key=lambda i: i[1], reverse=True)
            counter_morph_without_stopwords = []
            for word, num in counter_morph:
                if not word in stopwords_ko:
                    counter_morph_without_stopwords.append((word, num))

            one_result = pd.DataFrame([i,j,k,counter_morph_without_stopwords[:30]]).T
            if not len(result):
                result = one_result
            else:
                result = pd.concat([result, one_result])
result.reset_index(inplace=True)
result.drop('index',axis=1,inplace=True)
result.rename(columns={0:'',1:'',2:'_',3:f'most{most_words}_words'}, inplace=True)
result.to_csv(path+'_.csv', encoding='cp949')
display(result)
plt.figure(figsize=(40,100))
for i in range(len(result)):
    most30_words=result.iloc[i,-1]
    words = []
    values = []
    for j in most30_words:
        words.append(j[0])
        values.append(j[1])
    plt.subplot(24,2,i+1)
    plt.title(','.join(result.iloc[i,:-1].to_list()))
    plt.bar(range(most_words), values)
    plt.xticks(range(most_words), words, rotation=45)
plt.subplots_adjust(bottom=0.3, right=0.5, top=1, wspace=0.2, hspace=0.35)
plt.show()
```
