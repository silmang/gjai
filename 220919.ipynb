```python
import pandas as pd
import warnings
warnings.filterwarnings("error")
path = "C:\\pytest\\감성대화말뭉치\\"
```

```python
data = pd.read_csv(path+"감성대화말뭉치(최종데이터)_Validation.csv")
data = data[['연령','성별','감정_대분류','사람문장1','사람문장2','사람문장3','사람문장4']]
data
```

```python
data.info()
```

```python
temp_df1 = data[['연령','성별','감정_대분류','사람문장1']].rename(columns={'사람문장1':'사람문장'})
temp_df2 = data[['연령','성별','감정_대분류','사람문장2']].rename(columns={'사람문장2':'사람문장'})
temp_df3 = data[['연령','성별','감정_대분류','사람문장3']].dropna().rename(columns={'사람문장3':'사람문장'})
temp_df4 = data[['연령','성별','감정_대분류','사람문장4']].dropna().rename(columns={'사람문장4':'사람문장'})
```

```python
refined_data = pd.concat([temp_df1, temp_df2, temp_df3, temp_df4], axis=0)
refined_data.reset_index(inplace=True)
refined_data.drop('index', axis=1, inplace=True)
refined_data
```

```python
stopwords_ko = ["하다", "있다", "되다", "그", "않다", "없다", "나", "말", "사람", "이", "보다", "한", "때", "년", "같다", "대하다", "일", "이", "생각", "위하다", "때문", "그것", "그러나", "가다", "받다", "그렇다", "알다", "사회", "더", "그녀", "문제", "오다", "그리고", "크다", "속"]
most_words = 30
custom_stopwords_ko = ['나다','못','잘','다','많이','수','너무']

stopwords_ko = stopwords_ko+custom_stopwords_ko
result = pd.DataFrame()
for i in refined_data['연령'].unique():
    for j in refined_data['성별'].unique():
        for k in refined_data['감정_대분류'].unique():
            temp_df = refined_data.loc[(refined_data['연령']==i) & (refined_data['성별']==j) & (refined_data['감정_대분류']==k)]
            temp_sentences_list = temp_df['사람문장'].to_list()
            sentences = ' '.join(temp_sentences_list)
            morph = rhinoMorph.onlyMorph_list(rn, sentences, pos=['NNG', 'NNP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi=True)

            counter_morph = Counter(morph)
            counter_morph = sorted(counter_morph.items(), key=lambda i: i[1], reverse=True)
            counter_morph_without_stopwords = []
            for word, num in counter_morph:
                if not word in stopwords_ko:
                    counter_morph_without_stopwords.append((word, num))

            one_result = pd.DataFrame([i,j,k,counter_morph_without_stopwords[:30]]).T
            if not len(result):
                result = one_result
            else:
                result = pd.concat([result, one_result])
result.reset_index(inplace=True)
result.drop('index',axis=1,inplace=True)
result.rename(columns={0:'연령',1:'성별',2:'감정_대분류',3:f'most{most_words}_words'}, inplace=True)
result.to_csv(path+'연령성별감정별_키워드.csv', encoding='cp949')
display(result)
plt.figure(figsize=(40,100))
for i in range(len(result)):
    most30_words=result.iloc[i,-1]
    words = []
    values = []
    for j in most30_words:
        words.append(j[0])
        values.append(j[1])
    plt.subplot(24,2,i+1)
    plt.title(','.join(result.iloc[i,:-1].to_list()))
    plt.bar(range(most_words), values)
    plt.xticks(range(most_words), words, rotation=45)
plt.subplots_adjust(bottom=0.3, right=0.5, top=1, wspace=0.2, hspace=0.35)
plt.show()
```
