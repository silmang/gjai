```python
import pandas as pd

news_train = pd.read_csv(path+"뉴스데이터_train.csv", encoding='cp949')
news_test = pd.read_csv(path+"뉴스데이터_test.csv", encoding='cp949')
```

```python
# 분류2 제거
news_train.dropna(inplace=True)
news_test.dropna(inplace=True)
news_train.drop(['분류2','날짜','id'],axis=1,inplace=True)
news_test.drop(['분류2','날짜','id'],axis=1,inplace=True)
```

```python
def first_split(text):
    result=text.split(',')[0]
    if result=='정치' or result=='미분류':
        result=None
    return result
```

```python
news_train['분류1_상위']=news_train['분류1'].apply(first_split)
news_test['분류1_상위']=news_test['분류1'].apply(first_split)
```

```python
import rhinoMorph
rn = rhinoMorph.startRhino()

def morph(text):
    morphed = rhinoMorph.onlyMorph_list(
        rn, text, pos=['NNG', 'NNP', 'NP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'])
    return ' '.join(morphed)
```

```python
news_train['텍스트']=news_train['텍스트'].apply(morph)
news_test['텍스트']=news_test['텍스트'].apply(morph)
```

```python
import os, sys, glob
from PIL import Image #%pip install pillow
import numpy as np
from IPython.display import display
from IPython.display import Image as _Imgdis
import numpy as np
path = "C:\\pytest_img\\mnist_jpg\\"
```

```python
files = glob.glob(path+"*.jpg")
print("working with {0} images".format(len(files)))
```

```python
for i in range(2):
    print(files[i])
    display(_Imgdis(filename=files[i],width=30,height=40))
```

```python
from tensorflow.keras.preprocessing.image import array_to_img, img_to_array, load_img

file_names=[]
file_labels=[]

for _file in files:
    file_names.append(_file)
    file_labels.append(int(_file.split('_')[-2]))
```

```python
print(len(file_names))
print(len(file_labels))
set(file_labels)
```

```python
img = load_img(file_names[0])
print(type(img))

img_array = img_to_array(img)
print("img_array type:",type(img_array))
print("data type:",img_array.dtype)  # type: ignore
print("img shape:",img_array.shape)
```

```python
image_height=img_array.shape[0]
image_width=img_array.shape[1]
image_channels=img_array.shape[2]

dataset=np.ndarray(shape=(len(file_names),image_height,image_width,image_channels),dtype=np.int32)
print(dataset.shape)
```

```python
for count, item in enumerate(file_names):
    img=load_img(item)
    img_array=img_to_array(img)
    # img_array=img_array.reshape((28,28,3))
    dataset[count]=img_array
    if count%5000==0:
        print(f"{count} Images to array")
print("All images to array")
```

```python
print(dataset.shape)
print(file_labels[40000])
display(_Imgdis(filename=files[40000],width=30,height=40))
```

```python
np.set_printoptions(linewidth=np.inf)
print(dataset[40000,:,:,0])
```

```python
from sklearn.model_selection import train_test_split

train_images, test_images, train_labels, test_labels = train_test_split(dataset, file_labels, test_size=0.2)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
import numpy as np
```

```python
model=models.Sequential()
model.add(layers.Conv2D(32,kernel_size=(3,3),input_shape=(28,28,3),activation='relu'))
model.add(layers.Conv2D(64,kernel_size=(3,3),activation='relu'))
model.add(layers.MaxPool2D(pool_size=2))
model.add(layers.Dropout(0.25))
model.add(layers.Flatten())
model.add(layers.Dense(128,activation='relu'))
model.add(layers.Dropout(0.25))
model.add(layers.Dense(10,activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['acc'])
```

```python
train_images.shape
```

```python
train_images=train_images.reshape((len(train_images),28,28,3))
train_images=train_images.astype('float32')/255.
test_images=test_images.reshape((len(test_images),28,28,3))
test_images=test_images.astype('float32')/255.
```

```python
from tensorflow.keras.utils import to_categorical
train_labels=to_categorical(train_labels)
test_labels=to_categorical(test_labels)
```

```python
history=model.fit(train_images, train_labels, epochs=5, batch_size=128, validation_data=(test_images,test_labels))
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('test_acc:', test_acc)
```

```python
acc=history.history['acc']
val_acc=history.history['val_acc']
loss=history.history['loss']
val_loss=history.history['val_loss']
print('Accuracy of each epoch:',acc)
epochs=range(1,len(acc)+1)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,3))
plt.subplot(121)
plt.plot(epochs, acc, 'bo', label='Training Acc')
plt.plot(epochs, val_acc, 'b', label='Validation Acc')
plt.title("Training and validation accuracy")
plt.legend()
plt.subplot(122)
plt.plot(epochs, loss, 'bo', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title("Training and validation loss")
plt.legend()
plt.show()
```
