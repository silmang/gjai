```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(path+"apt.csv",encoding='cp949')
df.dropna(inplace=True)
df
```

```python
sns.set()
sns.barplot(x=df.id, y=df.Price)
plt.show()
```

```python
sns.set(context="notebook", style="white", palette='pastel')
sns.boxplot(x=df.Price)
plt.xlabel("apt price")
plt.show()
```

```python
df.Price
```

```python
sns.set(palette='dark')
sns.distplot(df.Price,color='red')
# sns.histplot(df.Price, kde=True, stat='density', color='red')
plt.xlabel('apt price')
plt.show()
```

```python
sns.set(palette='dark')
# sns.distplot(df.Price,color='red')
# sns.histplot(df.Price, kde=True, color='red', kde_kws=dict(cut=3))
sns.histplot(df.Price, kde=True, kde_kws=dict(cut=6))
plt.xlabel('apt price')
plt.show()
```

```python
sns.set(palette='dark')
# sns.distplot(df.Price,color='red')
sns.displot(df.Price, kde=True, color='red', kde_kws=dict(cut=3))
plt.xlabel('apt price')
plt.show()
```

```python
sns.set(context='talk')
sns.relplot(x='Price', y='FAR', data=df, color='green')
plt.xlabel('apt price')
plt.show()
```

```python
sns.set(context='poster', style='darkgrid')
sns.relplot(x='Price', y='FAR', data=df, color='red', kind='line')
plt.xlabel('apt price')
plt.show()
```

```python
sns.set()
sns.relplot(x='Price', y='FAR', data=df, kind='line')
plt.show()
```
