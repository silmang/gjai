```python
import pickle
def init_network():
    with open("c:\\pytest\\base\\sample_weight.pkl",'rb') as f:
        network=pickle.load(f)
    return network
def sigmoid(x):
    return 1/(1+np.exp(-x))
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y=exp_a/sum_exp_a
    return y
```

```python
def get_test_data():
    (train_image, train_label), (test_image, test_label) = load_mnist(flatten=True, normalize=True)
    return test_image, test_label
def predict(network, x):
    W1,W2,W3=network['W1'],network['W2'],network['W3']
    b1,b2,b3=network['b1'],network['b2'],network['b3']
    h1 = np.dot(x,W1)+b1
    o1 = sigmoid(h1)
    h2 = np.dot(o1,W2)+b2
    o2 = sigmoid(h2)
    out = np.dot(o2,W3)+b3
    y=softmax(out)
    return y
```

```python
images,labels = get_test_data()
network=init_network()
print(network)
```

```python
accuracy_cnt=0
for i in range(len(images)):
    y=predict(network, images[i])
    max_y=np.argmax(y)
    if max_y == labels[i]:
        accuracy_cnt+=1
print(f"accuracy: {str(float(accuracy_cnt)/len(images))}")
```

```python
batch_size=100
accuracy_cnt=0
for i in range(0, len(images), batch_size):
    x_batch=images[i:i+batch_size]
    y_batch=predict(network, x_batch)
    max_y_batch=np.argmax(y_batch,axis=1)
    accuracy_cnt+=np.sum(max_y_batch==labels[i:i+batch_size])
print(f"Accuracy: {str(float(accuracy_cnt)/len(images))}")
```

```python
import numpy as np
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y=exp_a/sum_exp_a
    return y
def cross_entropy_error(y,t):
    delta = .0000007
    return -np.sum(t*np.log(y+delta))
def numerical_gradient(f,x):
    h=0.0001
    grad=np.zeros_like(x)
    it=np.nditer(x,flags=['multi_index'],op_flags=['readwrite'])
    while not it.finished:
        index=it.multi_index

        tmp_val=x[index]
        x[index]=float(tmp_val)+h
        fxh1=f(x)

        x[index]=tmp_val
        fxh2=f(x)

        grad[index]=(fxh1-fxh2)/h
        it.iternext()
    return np.round(grad,3)
```

```python
class simpleNet:
    def __init__(self):
        self.W=np.random.randn(2,3)
    def predict(self,x):
        return np.dot(x,self.W)
    def loss(self, x,t):
        z=self.predict(x)
        y=softmax(z)
        loss=cross_entropy_error(y,t)
        return loss
```
