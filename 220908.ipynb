```python
import pandas as pd
path = "C:\\pytest_basic\\"
data = pd.read_csv(path+"bank.csv", header=0)
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=['job','education'], inplace=True)
data
```



열을 drop

```python
data.dropna(thresh=2170, axis=1)
```

```python
tmp = list(data.columns)
tmp.remove('poutcome')
data = data[tmp]
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.fillna({'contact':'unknown'})
data.head()
```

```python
data.isnull().sum()
```

```python
data = data[data.age >= 18]
data = data[data.age < 100]
data.shape
```

```python
data=data.replace('yes',1)
data=data.replace('no',0)
data.head()
```

```python
data = pd.get_dummies(data, columns=['job','marital','education','contact', 'month'])
data.head()
```

```python
data.to_csv(path+'result.csv',index=False)
```
