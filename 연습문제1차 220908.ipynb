```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
path="c:\\pytest_basic\\"
```

```python
test = pd.read_csv(path+'EC_X_test.csv')
X_train = pd.read_csv(path+'EC_X_train.csv')
y_train = pd.read_csv(path+'EC_y_train.csv')
```

```python
scale_1 = MinMaxScaler(); scale_1.fit(X_train[['Customer_care_calls']])
scale_2 = MinMaxScaler(); scale_2.fit(X_train[['Customer_rating']])
scale_3 = MinMaxScaler(); scale_3.fit(X_train[['Cost_of_the_Product']])
scale_4 = MinMaxScaler(); scale_4.fit(X_train[['Prior_purchases']])
scale_5 = MinMaxScaler(); scale_5.fit(X_train[['Discount_offered']])
scale_6 = MinMaxScaler(); scale_6.fit(X_train[['Weight_in_gms']])
```

```python
def preprocessing(test):
    try:
        test.drop(['ID'], axis=1, inplace=True)
    except:
        pass
    test['Gender'] = test['Gender'].replace('F', 0)
    test['Gender'] = test['Gender'].replace('M', 1)
    test_columns_list_onehot = list(test.columns)
    test_columns_list_onehot.remove('Customer_care_calls')
    test_columns_list_onehot.remove('Customer_rating')
    test_columns_list_onehot.remove('Cost_of_the_Product')
    test_columns_list_onehot.remove('Prior_purchases')
    test_columns_list_onehot.remove('Discount_offered')
    test_columns_list_onehot.remove('Weight_in_gms')
    test[['Customer_care_calls']] = scale_1.transform(test[['Customer_care_calls']])
    test[['Customer_rating']] = scale_2.transform(test[['Customer_rating']])
    test[['Cost_of_the_Product']] = scale_3.transform(test[['Cost_of_the_Product']])
    test[['Prior_purchases']] = scale_4.transform(test[['Prior_purchases']])
    test[['Discount_offered']] = scale_5.transform(test[['Discount_offered']])
    test[['Weight_in_gms']] = scale_6.transform(test[['Weight_in_gms']])
    test_one_hot_encoding = pd.get_dummies(test, columns=test_columns_list_onehot)
    return test_one_hot_encoding
```

```python
X_train_one_hot_encoding = preprocessing(X_train)
```

```python
X_train_one_hot_encoding
```

```python
model = RandomForestClassifier()
model.fit(X_train_one_hot_encoding, y_train['Reached.on.Time_Y.N'])
```

```python
pred = model.predict(X_train_one_hot_encoding)
```

```python
accuracy_score(y_train['Reached.on.Time_Y.N'], pred)
```

```python
# test 처리
test_one_hot_encoding = preprocessing(test)
```

```python
pred = model.predict(test_one_hot_encoding)
```

```python
result = pd.DataFrame([],columns=['id','Reached.on.Time_Y.N'])
result['id']= range(8800,11000)
result['Reached.on.Time_Y.N']=pred
result.to_csv(path+'EC_result.csv')
result
```
