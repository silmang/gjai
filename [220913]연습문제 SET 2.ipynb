### 연습문제 4

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
path = "c:\\pytest_basic\\"
train = pd.read_csv(path+"TravelInsurancePrediction_train.csv")
test = pd.read_csv(path+"TravelInsurancePrediction_test.csv")
```

```python
train_df = train.set_index('ID')
train_df.head()
```

```python
train_df.info() # 결측값이 없음
```

```python
train_df_train = train_df.iloc[:,:-1]
train_df_test = train_df.iloc[:,-1]
```

```python
train_df_train_num = train_df_train.select_dtypes(include='int64')
train_df_train_cat = train_df_train.select_dtypes(exclude='int64')
```

```python
scaler = MinMaxScaler()
scaler.fit(train_df_train_num)
scaled_array=scaler.transform(train_df_train_num)
train_df_train_num.iloc[:,:] = pd.DataFrame(scaled_array)
train_df_train_num
```

```python
train_df_train_cat= pd.get_dummies(train_df_train_cat)
train_df_train_cat
```

```python
model = RandomForestClassifier()
model.fit(pd.concat([train_df_train_num, train_df_train_cat], axis=1), train_df_test)
```

```python
pred = model.predict(pd.concat([train_df_train_num, train_df_train_cat], axis=1))
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(train_df_test, pred) # train dataset으로 93%
```

```python
# test
test_m = test.set_index('ID')

test_num = test_m.select_dtypes(include='int64')
test_cat = test_m.select_dtypes(exclude='int64')
testscaled_array=scaler.transform(test_num)
test_num.reset_index(inplace=True)
test_num.drop('ID', axis=1, inplace=True)
test_num.iloc[:,:] = pd.DataFrame(testscaled_array)
test_num
```

```python
test_cat.reset_index(inplace=True)
test_cat.drop('ID', axis=1, inplace=True)
test_cat= pd.get_dummies(test_cat)
test_cat
```

```python
pred = model.predict(pd.concat([test_num, test_cat], axis=1))
result = pd.concat([test[['ID']], pd.DataFrame(pred, columns=['predicated'])],axis=1)
result.set_index('ID')
```

```python
result.to_csv(path+'Travel_result_.csv', index=False)
```
